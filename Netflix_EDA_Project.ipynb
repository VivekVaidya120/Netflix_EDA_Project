```python
df['director'].fillna("Unknown",inplace=True)
df['cast'].fillna("Unknown",inplace=True)
df['country'].fillna("Unknown",inplace=True)
```

```python
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")
```

```python
df['rating'] = df['rating'].fillna("Unknown")
df['duration']= df['duration'].fillna("Unknown")
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
```

```python
df.isnull().sum()
```

```python
df['date_added'] = df['date_added'].fillna(pd.Timestamp("2000-01-01"))
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x='type',data=df,palette='Set2')
plt.title('Count of Movies and TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='type',data=df,hue='type',palette='Set2',legend=False)
plt.title('Count of Movies and TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df[['date_added','year_added']].head()                                                                                                                                                       
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='year_added',data=df,hue='year_added',palette='Set3',legend=False,order=sorted(df['year_added'].dropna().unique()))
plt.title('Content Added on Netflix by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values,y=top_countries.index,palette='muted')
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='rating',data=df,order=df['rating'].value_counts().index,palette='pastel')
plt.title('Distribution of Netflix Content by Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
df['listed_in'].value_counts().head(10)
```

```python
movie_df = df[df['type'] == 'Movie']
movie_df['duration'].value_counts().head(10).plot(kind='barh',color='coral')
plt.title('Top Movie Durations on Netflix')
plt.xlabel('Number of Movies')
plt.ylabel('Duration')
plt.show()
```

```python
df['director'].value_counts().head(10).plot(kind='barh', color='skyblue')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director Name')
plt.show()
```

```python
df['month_added'] = df['date_added'].dt.month
plt.figure(figsize=(10,5))
sns.countplot(x='month_added', data=df, palette='coolwarm')
plt.title('Content Added by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(ticks=range(0,12), labels=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.show()
```

```python
cast_df = df['cast'].dropna().str.split(',').explode().str.strip()
top_cast = cast_df.value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cast.values, y=top_cast.index, palette='Set1')
plt.title('Top 10 Most Frequent Actors on Netflix')
plt.xlabel('Number of Appearances')
plt.ylabel('Actor/Actress')
plt.show()
```

```python
content_over_time = df.groupby('date_added').size().resample('M').sum()
plt.figure(figsize=(14,6))
content_over_time.plot()
plt.title('Netflix Content Added Over Time (Monthly)')
plt.xlabel('Date')
plt.ylabel('Content Count')
plt.show()
```
